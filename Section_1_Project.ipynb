```python
# 결측치 제거
df_eda = df[df.notnull().all(axis=1)]

# 중복값 제거
df_eda = df_eda.drop_duplicates()
```

```python
# Year 타입 변환 (float -> int)
df_eda.Year = df_eda.Year.astype(int)

# Sales column들 타입 변환 (object -> float)
# df_eda.iloc[:, -4:] = df_eda.iloc[:, -4:].astype(float)

# 변환이 안되는 값이 있어서 체크
def notnum(x):
    try:
        float(x)
        return False
    except:
        return True

# Sales 부분만 분리 후
df_sales = df_eda.iloc[:, -4:]

# mask로 확인.
mask = df_sales.applymap(lambda x: notnum(x))
df_sales[mask.any(axis=1)]
```

```python
# Sales 값을 표준화. 
# M과 K가 있으니까 M 단위로 계산.

def convert_sales(sale):
  if sale[-1] == 'M':
    return float(sale[:-1]) * 1000000
  elif sale[-1] == 'K':
    return float(sale[:-1]) * 1000
  else:
    return float(sale) * 1000000
    
df_sales = df_sales.apply(lambda x: x.apply(convert_sales))
df_eda.update(df_sales)
```

```python
# 연도 이상치 확인
df_year = df_eda.query('(Year > 2050) or (Year < 1900)')

# 이상치의 분포 확인
plt.hist(df_year.Year)
```

```python
# 값을 확인해보니, 이상치의 분포가 0~20, 80~100 사이 구간에 있는걸 확인.
# 몇개의 외부 조사 한 결과, 0~20 구간은 2000년도, 80~100사이 구간은 1900년도로 추정됨.
# 이걸 근거로 이상치 변환 진행.

# 연도 이상치 
def convert_year(year):
  if year > 50:
    return year + 1900
  else:
    return year + 2000

df_year.Year = df_year.Year.apply(convert_year)
df_eda.update(df_year)
```

## **지역에 따라 선호하는 게임 장르가 다를까?**






### ❓ **어떻게 지역에 따른 선호도를 비교할 수 있을까?**




* **각 지역 별 가장 많이 판매된 게임 장르로 분석한다.**

  `판매량`은 각 지역 별로 해당 장르가 얼마나 '성공적' 이었는가? 라는 질문에 대한 답을 제공합니다.

  많은 게임 유저들이 관심을 가졌고, 적극적으로 구매에 동참했기에 판매량이 높게 집계되었을 것입니다.

  그렇기에, 판매량이 높은 데이터가 선호도가 높다고 판단할 수 있습니다.
  <br>

* **각 지역 별 가장 많이 출시된 게임 장르로 분석한다.**

  `출시량`은 각 지역 별로 얼마나 많은 게임사가 해당 게임에 관심을 가지고 있는지 보여주는 지표가 될 수 있습니다.

  다만 출시 된 게임이 오히려 유저들의 니즈와 상반되는 경우, 출시량에 비해 높은 판매량을 기록하지 못할 것입니다.

  따라서 출시 데이터를 반영한, '**게임 당 평균 판매량**' 을 보는게 더 바람직 할 수 있습니다.
  <br>

* **각 지역별 가장 평균 평점이 높은 게임 장르로 분석한다.**

  `평점`은 해당 장르에 대해 유저들이 어떻게 생각하는지 볼 수 있는 가장 직관적인 지표입니다.

  다만, 해당 데이터엔 평점에 관련된 항목이 포함되어있지 않으므로 해당 분석은 진행할 수 없습니다.

### ❓ **어떻게 연도별 게임의 트렌드의 유무를 확인할 수 있을까?**





* 연도별 판매량으로 비교

  `판매량` 으로 비교할 경우, 직관적으로 게임이 얼마나 많이 팔렸는가를 알 수 있기에 대략적인 인기나 트렌드를 확인하기 쉽습니다.

  다만, 시장의 사이즈가 커짐과 동시에 연도별 판매량이 같이 상승하기에, 얼핏 봐서는 모든 장르가 트렌드와 상관 없이 전부 잘 팔리는 것 처럼 느껴질 수 있습니다.
  <br>

* 연도별 판매율로 비교

  `판매율`로 비교할 경우, 시장의 사이즈와 상관 없이 조금 더 공정한 비교가 가능합니다.

  반대로, 시장이 작을 경우 판매율이 높아도 판매량이 기대에 못미칠 수 있습니다.
  <br>

따라서, 두 지표 모두 참고하며 연도별 게임의 트렌드를 분석합니다.

```python
# 연도별 데이터를 위한 변형
# 총 판매량이 높은 4개만 끊어볼까?

df_pivot = pd.pivot_table(df_eda, index = 'Year', columns = 'Genre', values='Global_Sales', aggfunc='sum').reset_index()

df_pivot['Year'] = pd.to_datetime(df_pivot['Year'], format='%Y')
df_pivot.set_index('Year', inplace=True)
df_pivot = df_pivot.resample('5Y').sum()

sorted_value = df_pivot.sum().sort_values(ascending=False)
df_pivot = df_pivot.loc[:, sorted_value.index]

sum_by_col = df_pivot.sum(axis=0)
cols = sum_by_col.nlargest(4).index

# 선택된 열(column)로 구성된 새로운 데이터프레임을 생성한다
cols_to_keep = df_pivot.columns[df_pivot.columns.isin(cols)]
new_df_pivot = df_pivot[cols_to_keep]

# 판매량 -> 판매율로 변경
year_sum = new_df_pivot.sum(axis=1)
df_pivot2 = new_df_pivot.div(year_sum, axis=0)

# # 편향성을 줄이기 위해 합계의 root나 log 로 나누기. 

# # 편향성 제거 강도. 값이 작을수록 판매율에 근접
# remove_strength = 1.5

# # root
# # year_sum = df_pivot.sum(axis=1)**(1/remove_strength)
# # log
# # year_sum = np.log2(df_pivot.sum(axis=1))

# df_pivot2 = df_pivot.div(year_sum, axis=0)
```
